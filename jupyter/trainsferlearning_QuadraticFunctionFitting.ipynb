**任務:** 基於trainfer_data.csv數據(這裏采用隨機生成一個二次函數)，建立mlp模型，在實現模型遷移學習:
1. 實現x對y的預測，可視化結果
1. 基於新數據基於trainfer_data2.csv數據(這裏采用隨機生成另一個二次函數)，對前模型進行二次訓練，對比模型訓練次數少的情況下的表現‘

- **備注:** 模型結構：mlp，兩個隱藏層，每層50個神經元，激活函數relu，輸出層激活函數linear，迭代次數100次  

```python
#生成二次函数的点
import numpy as np
import matplotlib.pyplot as plt

# make up data建立数据
np.random.seed(1)
x_data = np.linspace(start=-5, stop=5, num=100)[:, np.newaxis]  #水平轴-7~10
# print(x_data)
#np.random.shuffle(x_data)
# noise = np.random.normal(0, 0.1, x_data.shape)
# y_data = np.square(x_data) + noise
y_data = np.square(x_data)
print(y_data.shape)
```

```python
#define X and y
X = x_data
y = y_data
```

```python
# plot input data
plt.scatter(X, y)  #将数据绘制图一元二次函数的数据集点
plt.title('x vs y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#建立模型
from keras.models import Sequential
from keras.layers import Dense
model1 = Sequential()
model1.add(Dense(units=50,input_dim=1,activation="relu"))
model1.add(Dense(units=50,activation="relu"))
model1.add(Dense(units=1,activation="linear"))
model1.compile(optimizer="adam",loss="mean_squared_error")  
model1.summary()
```

```python
#train tne model
model1.fit(X,y,epochs=100)
```

```python
#make prediction
y_predict = model1.predict(X)
```

```python
#每训练100次 画一次图
fig2 = plt.figure()
plt.scatter(X, y_data)
plt.plot(X,y_predict,'g')
plt.title('y-x 100 predict')
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
model1.fit(X,y,epochs=100)
y_predict = model1.predict(X)
```

```python
fig3 = plt.figure()
plt.scatter(X, y_data)
plt.plot(X,y_predict,'g')
plt.title('y-x 200 predict')
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
model1.fit(X,y,epochs=200)
y_predict = model1.predict(X)
```

```python
fig4 = plt.figure()
plt.scatter(X, y_data)
plt.plot(X,y_predict,'g')
plt.title('y-x 400 predict')
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

训练400次 拟合得很好了
